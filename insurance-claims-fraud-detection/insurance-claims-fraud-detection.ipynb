```python
#Import libs

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

**Reading Data**

```python
vendor_data = pd.read_csv('/kaggle/input/insurance-claims-fraud-data/vendor_data.csv')
vendor_data.head()
```

```python
insurance_data = pd.read_csv('/kaggle/input/insurance-claims-fraud-data/insurance_data.csv')
insurance_data.head()
```

```python
employee_data = pd.read_csv('/kaggle/input/insurance-claims-fraud-data/employee_data.csv')
employee_data.head()
```

# 1

```python
merging_data = pd.merge(employee_data,insurance_data, on='AGENT_ID', how='left')
merging_data.info()
```

```python
f_d = pd.merge(merging_data, vendor_data, on='VENDOR_ID', how='left')
f_d.info()
```

# 2

```python
df =f_d.groupby('INSURANCE_TYPE')['CUSTOMER_ID'].count().reset_index()
df
```

```python
df = df.sort_values(by='CUSTOMER_ID', ascending=False).reset_index()
df.head(3)
```
